```python
# Importar librería Pandas
import pandas as pd

# Cargar el CSV a un DataFrame
df = pd.read_csv('/content/spotify-2024.csv', delimiter=',', encoding='unicode_escape')

# Mostrar la data
df
```

```python
# Mostrar las columnas
df.columns
```

```python
# Lista de columnas a eliminar
cols_to_drop = [
    'YouTube Views',
    'YouTube Likes',
    'TikTok Posts',
    'TikTok Likes',
    'TikTok Views',
    'YouTube Playlist Reach',
    'Apple Music Playlist Count',
    'AirPlay Spins',
    'SiriusXM Spins',
    'Deezer Playlist Count',
    'Deezer Playlist Reach',
    'Amazon Playlist Count',
    'Pandora Streams',
    'Pandora Track Stations',
    'Soundcloud Streams',
    'Shazam Counts',
    'TIDAL Popularity'
]

# Eliminar las columnas
df.drop(columns=cols_to_drop, inplace=True)

# Mostrar la data
df
```

```python
# Convertir la columna 'Release Date' a datetime
df['Release Date'] = pd.to_datetime(df['Release Date'])

# Filtrar las filas donde el año de lanzamiento sea 2024
df_2024 = df[df['Release Date'].dt.year == 2024]

# Mostrar el DataFrame filtrado
df_2024
```

```python
# Ordenar por mes y día de lanzamiento de forma ascendente
df_2024 = df_2024.sort_values(by=['Release Date'])

# Mostrar el DataFrame filtrado y ordenado
df_2024
```

```python
# Eliminar filas donde 'Spotify Streams' sea NaN
df_2024 = df_2024.dropna(subset=['Spotify Streams'])

# Mostrar el DataFrame resultante
df_2024
```

```python
# Convertir la columna 'Spotify Streams' a tipo numérico, manejando las comas
df_2024['Spotify Streams'] = df_2024['Spotify Streams'].astype(str).str.replace(',', '').astype(int)

# Mostrar el DataFrame resultante
df_2024
```

```python
df_2024 = df_2024.copy()

df_2024['Track'] = df_2024['Track'].astype('string')

print(df_2024['Track'].dtype)
```

```python
# Renombrar columnas para facilitar su uso
new_columns = []
for col in df_2024.columns:
    new_col = col.lower().replace(" ", "_")
    new_columns.append(new_col)

df_2024.columns = new_columns

df_2024
```

```python
# Importar librería Altair
import altair as alt

# Activar el renderer de Altair para Google Colab
alt.renderers.enable('colab')
```

```python
# Crear gráfico de barras con Altair
mark_bar = alt.Chart(df_2024).mark_bar().encode(
    x=alt.X('spotify_streams:Q', title='Reproducciones'),
    y=alt.Y('track:N', sort='-x', title='Nombre de la Canción'),
    tooltip=[
        alt.Tooltip('track:N', title='Canción'),
        alt.Tooltip('artist:N', title='Artista'),
        alt.Tooltip('spotify_streams:Q', title='Reproducciones', format=',')  # Tooltip con formato numérico
    ]
).properties(
    title="Canciones más reproducidas en Spotify en 2024"
)

# Mostrar gráfico
mark_bar
```

```python
# Guardar gráfico de barras como .html
mark_bar.save('mark_bar.html')
```

```python
# Contar el número de canciones por día y mes
# Agrupar por mes y día, manteniendo el conteo y una fecha representativa
songs_by_day_month = (
    df_2024.groupby([df_2024['release_date'].dt.month.rename('Mes'),
                     df_2024['release_date'].dt.day.rename('Día')])
    .agg(
        songs_count=('release_date', 'count'),
        release_date=('release_date', 'first')  # Retener una fecha representativa
    )
    .reset_index()
)

# Crear el gráfico de burbujas
bubble_chart = alt.Chart(songs_by_day_month).mark_circle().encode(
    x=alt.X('Día:O', title='Día del Mes'),
    y=alt.Y('Mes:O', title='Mes del Año'),
    size=alt.Size('songs_count:Q', title='Número de Canciones'),
    color=alt.Color('songs_count:Q', scale=alt.Scale(scheme='blues'), legend=alt.Legend(title="Número de Canciones")),
    tooltip=[alt.Tooltip('release_date:T', title='Fecha'),
             alt.Tooltip('songs_count', title="Canciones")]
).properties(
    title='Número de Canciones Lanzadas por Día y Mes en 2024'
)

bubble_chart
```

```python
# Guardar gráfico de burbujas como .html
bubble_chart.save('bubble_chart.html')
```

```python
# Eliminar filas donde 'Spotify Popularity' sea NaN
df_2024 = df_2024.dropna(subset=['spotify_popularity'])

df_2024
```

```python
# Crea gráfico boxplot que muestra la popularidad de las canciones en Spotify y las diferencia según si son canciones explícitas o no explícitas
mark_boxplot = alt.Chart(df_2024).mark_boxplot().encode(
    x=alt.X('explicit_track:N'),
    y=alt.Y('spotify_popularity:Q', title='Spotify Popularity'),
    color=alt.Color('explicit_track:N'),
    tooltip=['track', 'artist']
).properties(
    title='Distribución de Popularidad en Spotify según el Contenido Explícito',
    width=300,
    height=400
).interactive()

mark_boxplot
```

```python
# Guardar gráfico de boxplot como .html
mark_boxplot.save('mark_boxplot.html')
```

```python
# Filtrar los 10 artistas con más streams
top_10_artists = df_2024.groupby('artist', as_index=False).agg(
    total_streams=('spotify_streams', 'sum'),
    song_count=('track', 'count')  # Contar el número de canciones por artista
).nlargest(10, 'total_streams')  # Seleccionar los 15 artistas con más streams

# Crear el gráfico
mark_bar_artists = alt.Chart(top_10_artists).mark_bar().encode(
    x=alt.X('artist:N', sort='-y', title='Artist'),  # Ordenar por total de streams
    y=alt.Y('total_streams:Q', title='Total Spotify Streams'),
    color=alt.Color('song_count:Q', title='Number of Songs'),  # Color basado en la cantidad de canciones
    tooltip=['artist', 'total_streams', 'song_count']  # Incluir artista, streams y número de canciones en el tooltip
).properties(
    title='Top 10 Artistas por reproducciones en spotify y cantidad de canciones, 2024',
    width=800,
    height=400
).interactive()

mark_bar_artists
```

```python
# Guardar gráfico de barras de artistas como .html
mark_bar_artists.save('mark_bar_artists.html')
```

```python
# Crear el gráfico de dona
alt.Chart(top_10_artists).mark_arc(innerRadius=50).encode(
    theta=alt.Theta('total_streams:Q', title='Total Spotify Streams'),  # Tamaño del arco según los streams
    color=alt.Color(
        'artist:N',
        title='Artist',
        sort=alt.EncodingSortField(field='total_streams', order='descending')  # Ordenar por streams en orden descendente
    ),
    tooltip=['artist', 'total_streams', 'song_count']  # Mostrar información adicional
).properties(
    title='Top 10 Artistas por reproducciones en spotify y cantidad de canciones, 2024 (Gráfico de Dona)'
)
```
